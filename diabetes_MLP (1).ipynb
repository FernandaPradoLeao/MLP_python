# Carregando os Dados

```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.head()
```

# Escolhendo atributos

```python
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df['Outcome']
```

# Separando entre conjuntos de treinamento e teste

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state = 5, # Controla o embaralhamento aplicado aos dados antes de aplicar a divisão
                                                    test_size = 0.25, # Porcentagem de dados para test
                                                    stratify = y # Os dados serão divididos de forma estratificada
                                                    )
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(X_train), len(X_test)))
```

# Normalização

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

X_train = mm.fit_transform(X_train)
X_test = mm.fit_transform(X_test)
```

# Treinamento

```python
import tensorflow as tf
from IPython.display import clear_output

classifier = tf.keras.models.Sequential()

# Primeira camada oculta com 64 neurônios e função de ativação 'relu'.
# A camada de entrada tem o mesmo número de neurônios que o número de características (8).
classifier.add(tf.keras.layers.Dense(64, activation='relu', input_shape=(8,)))
# Segunda camada oculta
classifier.add(tf.keras.layers.Dense(32, activation='relu'))
# Camada de saída com ativação 'sigmoid' para classificação binária.
classifier.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Compilar o modelo
# Usamos o otimizador 'adam' e a função de perda 'binary_crossentropy' para classificação binária.
classifier.compile(optimizer="adam", loss = "binary_crossentropy", metrics= ['accuracy'] )

# Treinamento do modelo
# Usamos X_train e y_train com um número definido de épocas.
classifier.fit(X_train, y_train, batch_size = 1, epochs = 100)

clear_output(wait=True)
```

# Avaliando o modelo

```python
import numpy as np
from sklearn.metrics import classification_report

# 'y_pred' contém as probabilidades preditas pela rede neural
y_pred = classifier.predict(X_test)

# Converter probabilidades em rótulos de classe 0 ou 1
# Usamos um limiar de 0.5
y_pred_bin = [ ( 1 if elem > 0.5 else 0) for elem in y_pred ]

# Transforma em array
y_pred_bin =  np.array(y_pred_bin)

print(classification_report(y_test, y_pred_bin))
```

```python
from sklearn.metrics import accuracy_score
print("Acurácia: ", round(accuracy_score(y_test,y_pred_bin)*100,2), "%"  )
```
